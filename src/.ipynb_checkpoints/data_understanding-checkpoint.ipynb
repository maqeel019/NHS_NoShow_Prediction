```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("../data/processed/nhs_no_show_enhanced_synthetic.csv")
df
```

```python
df.info()
df.describe()
df.isnull().sum()
```

```python
df.describe(include='all').T
```

```python
print("Missing values per column:")
print(df.isnull().sum())

# If any duplicate rows remain, drop them now
print('*'*30)
print("Duplicates before:", df.duplicated().sum())
df = df.drop_duplicates().reset_index(drop=True)
print("Duplicates after:", df.duplicated().sum())
print('*'*30)

```

```python

print(df['Gender'].value_counts(normalize=True) * 100)

print(df['IMD_Decile'].value_counts(normalize=True) * 100)

df['Age_Group'].value_counts().sort_values(ascending=False)
```

```python
noshow_rate = df['NoShow'].value_counts(normalize=True).mul(100)
print("No-show rate summary (%):\n", noshow_rate)

```

```python
df.groupby('Gender')['NoShow'].value_counts(normalize=True).unstack() * 100
```

```python

df.groupby("Age_Group")['NoShow'].value_counts(normalize=True).unstack() * 100
```

```python
df.groupby('IMD_Decile')['NoShow'].value_counts(normalize=True).unstack() * 100

```

```python

df.groupby('Consultation_Type')['NoShow'].value_counts(normalize=True).unstack() * 100
```

```python
df.groupby('Appointment_Type')['NoShow'].value_counts(normalize=True).unstack() * 100
```

```python
# Group by and count
df = df[df['Medical_Specialty'] != 'Total Activity']

specialty_counts = df.groupby(['Medical_Specialty', 'Gender', 'Age_Group']).size().reset_index(name='Count')

# Sort by count descending
specialty_counts = specialty_counts.sort_values(by='Count', ascending=False).reset_index(drop=True)

# Display
specialty_counts.head(10)
```
